```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Creating the dataset, removing null and duplicates.
df_spotify = pd.read_csv('https://raw.githubusercontent.com/MateusViotto/Machine-Learning-with-Python/main/Datasets/Spotify-dataset.csv', index_col=0).dropna().drop_duplicates()

df_spotify.head()
```

```python
#Converting milliseconds to seconds of track duration

df_spotify["duration_ms"] = df_spotify["duration_ms"] / 1000

df_spotify.rename(columns={"duration_ms":"duration_sec"},inplace=True)

df_spotify.head()
```

```python
df_spotify.describe()
```

### Converting Popularity column to binary.

```python
conditions = [(df_spotify['popularity'] >= 70), (df_spotify['popularity'] < 70)]

values = [1,0]

df_spotify['pop_bool'] = np.select(conditions,values,default='Not Specified')

df_spotify['pop_bool'] = df_spotify['pop_bool'].astype(int)

#Popularity is now converted to a binary value, where if popularity has a value greater or equal to 70 is now considered as True. 

df_spotify.head()
```

### Creating df with only quantitative values

```python
df_quantSpotify = df_spotify.copy()

df_quantSpotify = df_quantSpotify.drop(columns = ['track_id','artists', 'album_name','track_name','explicit','track_genre','popularity'])

df_quantSpotify.head()
```

### Normalizing Data

```python
df_quantSpotify_nm = (df_quantSpotify-df_quantSpotify.min())/(df_quantSpotify.max()-df_quantSpotify.min())

df_quantSpotify_nm.head()
```

---

### KFold Cross Validation

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
```

```python
x = df_quantSpotify_nm.drop('pop_bool', axis=1)

y = df_quantSpotify_nm['pop_bool']
```

```python
model = LinearRegression()
kfold = KFold(n_splits=10, shuffle=True)
result = cross_val_score(model, x, y, cv = kfold)

print(result.mean()*100)
```

```python
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
linearRegression = LinearRegression()
elasticNet       = ElasticNet()
ridge            = Ridge()
lasso            = Lasso()

  # Applyes KFold to models.
linearRegression_result = cross_val_score(linearRegression, x, y, cv = kfold)
elasticNet_result       = cross_val_score(elasticNet, x, y, cv = kfold)
ridge_result            = cross_val_score(ridge, x, y, cv = kfold)
lasso_result            = cross_val_score(lasso, x, y, cv = kfold)
```

```python
dic_models = {
    "LinearRegression": linearRegression_result.mean(),
    "ElasticNet": elasticNet_result.mean(),
    "Ridge": ridge_result.mean(),
    "Lasso": lasso_result.mean()
}
  # Select the best model.
bestModel = max(dic_models, key=dic_models.get)

print("Linear Regression Mean (R^2): {0}\nElastic Net Mean (R^2): {1}\nRidge Mean (R^2): {2}\nLasso Mean (R^2): {3}".format(linearRegression_result.mean(), elasticNet_result.mean(), ridge_result.mean(), lasso_result.mean()))
print("The best model is: {0} with value: {1}".format(bestModel, dic_models[bestModel]))
```

### Alura

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
```

```python
SEED = 5
np.random.seed(SEED)
treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, test_size = 0.25,stratify = y)
print("Treinaremos com %d elementos e testaremos com %d elementos" % (len(treino_x), len(teste_x)))
```

```python
from sklearn.dummy import DummyClassifier

dummy_stratified = DummyClassifier(strategy='stratified')
dummy_stratified.fit(treino_x, treino_y)
acuracia = dummy_stratified.score(teste_x, teste_y) * 100

print("A acurácia do dummy stratified foi de %.2f%%" % acuracia)
```

```python
from sklearn.tree import DecisionTreeClassifier


modelo = DecisionTreeClassifier(max_depth=2)
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print ("A acurácia foi %.2f%%" % acuracia)
```
