```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Creating the dataset, removing null and duplicates.
df_spotify = pd.read_csv('https://raw.githubusercontent.com/MateusViotto/Machine-Learning-with-Python/main/Datasets/Spotify-dataset.csv', index_col=0).dropna().drop_duplicates()

df_spotify.head()
```

```python
#Converting milliseconds to seconds of track duration

df_spotify["duration_ms"] = df_spotify["duration_ms"] / 1000

df_spotify.rename(columns={"duration_ms":"duration_sec"},inplace=True)

df_spotify.head()
```

```python
df_spotify.describe()
```

# First Analysis

### Top 100 Tracks

```python
df_toptracks = df_spotify.sort_values('popularity',ascending = False).head(100)

df_toptracks
```

```python
df_toptracks_view = df_toptracks[['track_name', 'popularity']].copy()

df_toptracks_view.rename(columns={"track_name":"Tracks","popularity":"Popularity"},inplace=True)

df_toptracks_view
```

### Most Popular Artists

```python
# Assuming your dataframe is named 'df' with columns 'artist' and 'popularity'
df_grouped_artist = df_spotify.groupby('artists')['popularity'].sum().reset_index()

# Sort the dataframe based on the sum of popularity in descending order
df_artist_sorted = df_grouped_artist.sort_values(by='popularity', ascending=False)

# Display the most popular artists
most_popular_artists = df_artist_sorted['artists']
most_popular_artists.head(100)
```

### Most Popular Genres

```python
# Assuming your dataframe is named 'df' with columns 'artist' and 'popularity'
df_grouped_genres = df_spotify.groupby('track_genre')['popularity'].sum().reset_index()

# Sort the dataframe based on the sum of popularity in descending order
df_genres_sorted = df_grouped_genres.sort_values(by='popularity', ascending=False)

display(df_genres_sorted)

# Display the most popular artists
most_popular_genres = df_genres_sorted['track_genre']
most_popular_genres.head(100)
```

### Musics by Artists

```python
df_tracksbyArtists = df_spotify["artists"].value_counts()

print(df_tracksbyArtists[:].sum())

df_tracksbyArtists

```

### Musics by Genre

```python
df_tracksbyGenre = df_spotify["track_genre"].value_counts()

df_tracksbyGenre
```

### Top 10 Longest Tracks

```python
df_longestTracks = df_spotify[["track_name","duration_sec"]].sort_values(by="duration_sec",ascending=False)[:10]

df_longestTracks
```

### Top 10 Most Danceable Tracks

```python
df_mostDanceableTracks = df_spotify[["track_name","artists","danceability"]].sort_values(by="danceability",ascending=False)[:10]

df_mostDanceableTracks
```

## Correlation Heatmap

```python
df_corr = df_spotify.corr(method="pearson",numeric_only=True)

df_corr
```

```python
#Ploting correlation table using Seaborn
plt.figure(figsize=(16,5))
sns.heatmap(df_corr, annot=True, fmt=".1g")
plt.title("Correlation Heatmap between variables")
plt.show()
```

### Converting Popularity column to binary.

```python
conditions = [(df_spotify['popularity'] >= 70), (df_spotify['popularity'] < 70)]

values = [1,0]

df_spotify['pop_bool'] = np.select(conditions,values,default='Not Specified')

df_spotify['pop_bool'] = df_spotify['pop_bool'].astype(int)

#Popularity is now converted to a binary value, where if popularity has a value greater or equal to 70 is now considered as True. 

df_spotify.head()
```

### Creating df with only quantitative values

```python
df_quantSpotify = df_spotify.copy()

df_quantSpotify = df_quantSpotify.drop(columns = ['track_id','artists', 'album_name','track_name','explicit','track_genre','popularity'])

df_quantSpotify.head()
```

### Normalizing Data

```python
df_quantSpotify_nm = (df_quantSpotify-df_quantSpotify.min())/(df_quantSpotify.max()-df_quantSpotify.min())

df_quantSpotify_nm.head()
```
