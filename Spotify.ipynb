```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_spotify = pd.read_csv('https://raw.githubusercontent.com/MateusViotto/Machine-Learning-with-Python/main/Datasets/Spotify-dataset.csv').dropna()

df_spotify.head()
```

# First Analysis

### Top 100 Tracks

```python
df_toptracks = df_spotify.sort_values('popularity',ascending = False).head(100)

df_toptracks
```

```python
df_toptracks_view = df_toptracks[['track_name', 'popularity']].copy()

df_toptracks_view.rename(columns={"track_name":"Tracks","popularity":"Popularity"},inplace=True)

df_toptracks_view
```

### Most Popular Artists

```python
# Assuming your dataframe is named 'df' with columns 'artist' and 'popularity'
df_grouped_artist = df_spotify.groupby('artists')['popularity'].sum().reset_index()

# Sort the dataframe based on the sum of popularity in descending order
df_artist_sorted = df_grouped_artist.sort_values(by='popularity', ascending=False)

# Display the most popular artists
most_popular_artists = df_artist_sorted['artists']
most_popular_artists.head(100)
```

### Most Popular Genres

```python
# Assuming your dataframe is named 'df' with columns 'artist' and 'popularity'
df_grouped_genres = df_spotify.groupby('track_genre')['popularity'].sum().reset_index()

# Sort the dataframe based on the sum of popularity in descending order
df_genres_sorted = df_grouped_genres.sort_values(by='popularity', ascending=False)

display(df_genres_sorted)

# Display the most popular artists
most_popular_genres = df_genres_sorted['track_genre']
most_popular_genres.head(100)
```
